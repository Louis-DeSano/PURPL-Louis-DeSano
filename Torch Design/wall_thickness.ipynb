```python
sigma_yield = 30000 #psi
temp_factor = 0.55 # percent reduction in strength due to temperature
sigma_temp = sigma_yield * temp_factor # yield stress at temp
```

```python
# approximate as thick walled cylinder
p = 500 #pressure [psi]

r_o = 0.4 # actual thickness [in] (approximate as the outer radius of a cylinder)
r_i = 0.172 # radius [in]

sigma = p*(r_o**2 + r_i**2) / (r_o**2 - r_i**2)
FOS_actual =  sigma_temp / sigma
print(f"Design Thickness {r_o:0.3f} in")
print(f"Design Stress: {sigma:0.3f} psi")
print(f"Design FOS: {FOS_actual:0.3f}")
```
